```python
import numpy as np
import math
import string
from sympy import Matrix, pprint
```

```python
key = input()
key = list(map(int,key.split(' ')))
```

```python
n = math.sqrt(len(key))
n = int(n)
key = np.array(key).reshape(n,n)
key
```

```python
raw_text = input()
```

```python
def prepare_text(raw_text):
    input_text = list(raw_text)
    input_text = list(filter((' ').__ne__, input_text))
    text = np.zeros(len(input_text))
    for a in range(len(input_text)):
        text[a] = ord(input_text[a]) - ord('a')
    print(text)
    l = len(input_text)
    print(l)
    if(l%n !=0 ):
        a = np.ones(n-l%n)*25
        print(a)
        text = np.append(text,a,axis = 0)
    text = text.reshape(int(text.shape[0]/n),n)
    return text
```

```python
def encrypt(raw_text):
    text = prepare_text(raw_text)
    enc = []
    enc_text = ""
    for a in range(len(text)):
        b = np.dot(key,text[a])
        for i in range(len(b)):
            enc += [b[i]%26]
    print(enc)
    ind = 0
    for i in range(len(raw_text)):
        if(raw_text[i]!=' '):
            enc_text += chr(ord('a') + int(enc[ind])) 
            ind +=1
        else:
            enc_text += ' '
    return enc_text

```

```python
encrypt(raw_text)
```

```python
raw_cipher = input()
```

```python
def decrypt(raw_cipher):
    cipher = prepare_text(raw_cipher)
    tempkey = Matrix(key)
    kinv = tempkey.inv_mod(26)
    dec = []
    dec_text = ""
    for a in range(len(cipher)):
        b = np.dot(kinv,cipher[a])
        for i in range(len(b)):
            dec += [b[i]%26]
    print(dec)
    ind = 0
    for i in range(len(raw_cipher)):
        if(raw_cipher[i]!=' '):
            dec_text += chr(ord('a') + int(dec[ind])) 
            ind +=1
        else:
            dec_text += ' '
    return dec_text
    
```

```python
decrypt(raw_cipher)
```
